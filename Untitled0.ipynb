```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("shaunthesheep/microsoft-catsvsdogs-dataset")

print("Path to dataset files:", path)
```

```python
pet_images_path = os.path.join(path, 'PetImages')

for folder in os.listdir(pet_images_path):
    folder_path = os.path.join(pet_images_path, folder)
    if os.path.isdir(folder_path):
        count = len(os.listdir(folder_path))
        print(f"{folder}: {count} images")
```

```python
pet_images_path = os.path.join(path, 'PetImages')
cat_dir = os.path.join(pet_images_path, 'Cat')
dog_dir = os.path.join(pet_images_path, 'Dog')
```

```python
cat_images = [(os.path.join(cat_dir, f), 0) for f in os.listdir(cat_dir)]  # 0 = Cat
dog_images = [(os.path.join(dog_dir, f), 1) for f in os.listdir(dog_dir)]  # 1 = Dog
all_images = cat_images + dog_images

print(f" الصور: {len(all_images)}")
print(f" صور القطط: {len(cat_images)}")
print(f" صور الكلاب: {len(dog_images)}")
```

```python
random.shuffle(all_images)
```

```python
image_paths = [x[0] for x in all_images]
labels = [x[1] for x in all_images]
```

```python
train_paths, temp_paths, train_labels, temp_labels = train_test_split(
    image_paths, labels, test_size=0.30, random_state=42
)


val_paths, test_paths, val_labels, test_labels = train_test_split(
    temp_paths, temp_labels, test_size=0.50, random_state=42
)
```

```python
print(f" Train:      {len(train_paths)} صورة")
print(f" Validation: {len(val_paths)} صورة")
print(f" Test:       {len(test_paths)} صورة")
```

```python
IMG_SIZE = (128, 128)
BATCH_SIZE = 32
```

```python
def load_images(paths, labels):
    images = []
    valid_labels = []
    for p, l in zip(paths, labels):
        try:
            img = load_img(p, target_size=IMG_SIZE)   # تغيير الحجم
            img = img_to_array(img) / 255.0            # Normalize
            images.append(img)
            valid_labels.append(l)
        except:
            pass
    return np.array(images), np.array(valid_labels)
```

```python
X_train, y_train = load_images(train_paths, train_labels)

X_val, y_val = load_images(val_paths, val_labels)

X_test, y_test = load_images(test_paths, test_labels)

print(f"X_train shape: {X_train.shape}")
print(f"X_val shape:   {X_val.shape}")
print(f"X_test shape:  {X_test.shape}")
```

```python
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
fig.suptitle('Sample Images from Dataset', fontsize=16)

cat_indices = np.where(y_train == 0)[0][:5]
for i, idx in enumerate(cat_indices):
    axes[0, i].imshow(X_train[idx])
    axes[0, i].set_title(' Cat')
    axes[0, i].axis('off')

dog_indices = np.where(y_train == 1)[0][:5]
for i, idx in enumerate(dog_indices):
    axes[1, i].imshow(X_train[idx])
    axes[1, i].set_title(' Dog')
    axes[1, i].axis('off')

plt.tight_layout()
plt.show()
```

```python
from tensorflow.keras import layers, models

model = models.Sequential([

    layers.Conv2D(32, (3,3), activation='relu', input_shape=(128, 128, 3)),
    layers.MaxPooling2D(2, 2),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2, 2),

    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D(2, 2),

    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(1, activation='sigmoid')  # Binary: Cat or Dog
])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
early_stop = EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
checkpoint = ModelCheckpoint('best_model.keras',monitor='val_accuracy',save_best_only=True, verbose=1)


trainnig = model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=64,
    validation_data=(X_val, y_val),
    callbacks=[early_stop, checkpoint]
)
```

```python
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int).flatten()

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Cat ', 'Dog '],
            yticklabels=['Cat ', 'Dog '])
plt.title('Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Loss
axes[0].plot(trainnig.history['loss'], label='Train Loss', color='blue')
axes[0].plot(trainnig.history['val_loss'], label='Val Loss', color='red')
axes[0].set_title('Model Loss')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Loss')
axes[0].legend()

# Accuracy
axes[1].plot(trainnig.history['accuracy'], label='Train Accuracy', color='blue')
axes[1].plot(trainnig.history['val_accuracy'], label='Val Accuracy', color='red')
axes[1].set_title('Model Accuracy')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Accuracy')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
print(f" Accuracy:  {accuracy_score(y_test, y_pred):.4f}")
print(f" Precision: {precision_score(y_test, y_pred):.4f}")
print(f" Recall:    {recall_score(y_test, y_pred):.4f}")
print(f" F1-Score:  {f1_score(y_test, y_pred):.4f}")
print(" Classification Report:")
print(classification_report(y_test, y_pred, target_names=['Cat', 'Dog']))
```

```python
from tensorflow.keras.utils import load_img, img_to_array
import numpy as np
import matplotlib.pyplot as plt

def predict_image(img_path):
    # تحميل وتحضير الصورة
    img = load_img(img_path, target_size=(128, 128))
    img_array = img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)  #  batch dimension

    # التنبؤ
    prediction = model.predict(img_array)[0][0]

    # النتيجة
    label = " Dog" if prediction > 0.5 else " Cat"
    confidence = prediction if prediction > 0.5 else 1 - prediction

    plt.figure(figsize=(5, 5))
    plt.imshow(load_img(img_path))
    plt.title(f"Prediction: {label}\nConfidence: {confidence:.2%}", fontsize=14)
    plt.axis('off')
    plt.show()

    print(f" result: {label}")
    print(f" confidence: {confidence:.2%}")

predict_image("/Users/ahadm/Desktop/cv/cnn-classification-ahad-m/image.png")
predict_image("/Users/ahadm/Desktop/cv/cnn-classification-ahad-m/image copy.png")

```
